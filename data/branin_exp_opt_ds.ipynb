```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import torch
from sklearn.model_selection import train_test_split
from pathlib import Path
```

```python
res_dir = Path('f_opt').joinpath('branin')
res_dir.mkdir(exist_ok=True,parents=True)
```

```python
N = 1000
x1 = np.linspace(-5, 10, N)
x2 = np.linspace(0, 15, N)
X0, X1 = np.meshgrid(x1, x2)

Y = np.zeros(N)
Y = np.nan
Y = (X1 - 5.1 / (4 * np.pi ** 2) * X0 ** 2 + 5 / np.pi * X0 - 6) ** 2 + 10 * (1 - 1 / (8 * np.pi)) * np.cos(X0) + 10

y = np.zeros(N)
y = np.nan
y = (x2 - 5.1 / (4 * np.pi ** 2) * x1 ** 2 + 5 / np.pi * x1 - 6) ** 2 + 10 * (1 - 1 / (8 * np.pi)) * np.cos(x1) + 10
#x_minor, _, y_minor, _ = train_test_split(x,y, train_size=0.01)

#df = pd.read_csv(res_dir.joinpath('synth_full_ds_0.csv'))
#x_minor = df[['x1','x2','x3']].to_numpy()
#y_minor = df['y'].to_numpy()

plt.figure(figsize=(8, 6))
contour = plt.contourf(X0, X1, Y, levels=50, cmap='viridis')  # заливка
contour_lines = plt.contour(X0, X1, Y, levels=200, colors='black', linewidths=0.5)  # линии
#plt.scatter(x_minor.T[0], y_minor)
plt.colorbar(contour)
plt.title("Контурный график функции Брэнина")
plt.xlabel("x0")
plt.ylabel("x1")

plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig(res_dir.joinpath('bronin_contour.png'))
```
